```python
# check if a number is prime
def isz_prime(n):
    for i in range (2,n,1):
        if n%i==0:
            print(n, 'is not prime')
            break
        else:
            print (n, 'is prime')
            break
        
# now call the function
for r in range(2,1001,1):
    isz_prime(r)
```
