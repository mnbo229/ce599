```python
# check if a number is prime
def is_prime(n):
    for i in range (2,n,1):
        if n%i==0:
            print(n, 'is not prime')
            break
        else:
            print (n, 'is prime')
        
# now call the function
#is_prime(80)

#now repeat from 2 to 1000:
for r in range (0,1001,1):
    is_prime(r)
```
